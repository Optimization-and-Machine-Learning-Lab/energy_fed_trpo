```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_graph(b1, b2, index):
    file_path = "/home/yunxiang.li/FRL/CityLearn/citylearn/data/my_data/building_simple_"+str(b1)+".csv"
    file_path_test = "/home/yunxiang.li/FRL/CityLearn/citylearn/data/my_data/building_simple_"+str(b2)+".csv"

    demand = []
    with open(file_path, "r") as f:
        csv_reader = csv.reader(f, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                # print(f'Column names are {", ".join(row)}')
                line_count += 1
            else:
                # print(row[-5])  consumption: -5, solar: -1
                demand.append(float(row[index]))

    demand_test = []
    with open(file_path_test, "r") as f:
        csv_reader = csv.reader(f, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0:
                # print(f'Column names are {", ".join(row)}')
                line_count += 1
            else:
                # print(row[-5])  consumption: -5, solar: -1
                demand_test.append(float(row[index]))

    plot_demand_train = []
    plot_demand_test = []
    day = np.array(list(range(0, 24)))
    for i in range(3, 9):
        for j in range(30):
            plot_demand_train.append(demand[(i-1)*30*24+j*24: (i-1)*30*24+j*24+24])
    for j in range(30):
        plot_demand_test.append(demand_test[j*24: j*24+24])

    plot_demand_train = np.array(plot_demand_train)
    plot_demand_test = np.array(plot_demand_test)

    for i in range(len(plot_demand_train)):
        plt.plot(day, plot_demand_train[i], 'r-')
        # print(plot_demand_train[i-4])
    for i in range(len(plot_demand_test)):
        plt.plot(day, plot_demand_test[i], 'b--')
    plt.title("train_"+str(b1)+"_test_"+str(b2))
    plt.show()
```

```python
for i in range(5):
    for j in range(5):
        draw_graph(i+1, j+1, -1)    # solar generation
```

```python
for i in range(5):
    for j in range(5):
        draw_graph(i+1, j+1, -5)    # demand
```
